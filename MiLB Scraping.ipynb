```python
# Dependencies
import pandas as pd
import requests as rs
from bs4 import BeautifulSoup as bs
from splinter import Browser
from splinter.exceptions import ElementDoesNotExist
import time
```

```python
# Base search URL for the basebaseballcube website with a minimum of 30 plate appearances.
burl = 'http://www.thebaseballcube.com/minors/stats.asp?Q=Y&Report=Batting&D1=&D2=&R=&R1==&dteam=&minPA=30&Player=&RL=T&Lg=&Sort=PA&Pos=&Team=&Conditions='
```

```python
#Lists for cycling through with years, organizations, and levels


years = list(range(1977,1994)) 
# years = [1977, 1978]
orgs = list(range(2, 33))
# orgs = list(range(2, 3))
levels = ['AAA', 'AA']
# levels = ['AAA', 'AA']
```

```python
#Cycles through every year, team, and level
final = pd.DataFrame()
for year in years:
    for org in orgs:
        for level in levels:
            #Reads all the tables on the webpage
            test = pd.read_html(f'{burl}&Y1={year}&Y2={year}&Lev={level}&Org={org}')
            #The search results end up as the 18th table on the page
            test = test[18]
            # The first column is the labels of the columns, so we replace the column names with the first row
            test.columns = test.iloc[0]
            # And then we save everything but the first row, and append it to our single database
            test = test.iloc[1:]
            final = final.append(test)
```

```python
#Each individual scrape that is appended has 0-25ish index, so we reset index

final = final.reset_index()

#This creates an 'index' column with the old value, which we delete
final = final.drop(columns=['index'])
```

```python
# Saves the data
final.to_csv("temp.csv")
```
